```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
np.random.seed(42)
```

```python
ad_spend = np.random.uniform(5, 50, 100)
```

```python
noise = np.random.normal(0, 5, 100)
data_sales = 3.5 * ad_spend + noise
```

```python
data = pd.DataFrame({
    'Advertising_Spend': ad_spend,
    'Sales': data_sales
})
```

```python
data.to_csv('business_linear_regression_dataset.csv', index=False)
print("Dataset saved as 'business_linear_regression_dataset.csv'")
```

```python
df = pd.read_csv('business_linear_regression_dataset.csv')
df
```

```python
plt.scatter(df['Advertising_Spend'],df['Sales'])
plt.xlabel('Advertising_Spend')
plt.ylabel('Sales(in rupees)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
plt.scatter(df['Advertising_Spend'],df['Sales'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Advertising_Spend')
plt.ylabel('Sales(in rupees)')
```
